```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)       
Xa=scaler.fit_transform(X) 
from sklearn.model_selection import train_test_split
X_train_std, X_test_std, y_train, y_test = train_test_split(Xa, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100) 
pca.fit(X_train_std)
```

```python
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
```

```python
X_test_pca
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca.components_
```

```python
from sklearn.linear_model import Lasso
model_lasso2 = Lasso(alpha=1000)                                
model_lasso2.fit(X_train_pca, y_train) 
```

```python
model_lasso2.coef_.shape
```

```python
model_lasso.coef_
```

```python
model_lasso2.score(X_test_pca,y_test)
```
